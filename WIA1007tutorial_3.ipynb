```python
import numpy as np
array1 = np.array([[1,2],
                   [3,4]])
print("1. 2*2 numpy array:\n", array1)
```

```python
array2 = np.random.rand(3,4)
print("\n2. Random numpy array (3*4):\n", array2)
```

```python
ones_array = np.ones((3,3))
zeroes_array = np.zeros((3,3))
print("\n3. Array of ones:\n", ones_array)
print("Array of zeroes:\n", zeroes_array)
```

```python
array3 = np.random.rand(6,3)
print("\n4. Random array (6*3):\n", array3)
print("Transposed array (3*6):\n", array3.transpose)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)

print("\n5. Matrix A (2*3):\n", A)
print("Matrix B (3*4):\n", B)
print("Dot product (2*4):\n", dot_product)
```
